The task is to create a decision tree classifier and visualize it graphically

Purpose - If we feed any new data to the classifier, it would be able to predict the right class accordingly

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import sklearn.datasets as datasets
import seaborn as sns
```

```python
# Loading the iris dataset
df = pd.read_csv("/content/Iris.csv")
df['Species_label'], _ = pd.factorize(df['Species'])
df.head()
```

```python
iris = datasets.load_iris()
y = iris.target
print(y)
```

```python
# Proper distributed data
df.describe().T
```

```python
df.isna().sum().sum()
```

**Splitting the data**

```python
# Defining the decision tree algorithm
x = df.iloc[:,1:5]
y = df.iloc[:,-1]

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1010)
clf = DecisionTreeClassifier(random_state=0)
clf.fit(x_train,y_train)
```

```python
# Prediction of trained data and accuracy
pred = clf.predict(x_test)
print(pred)
from sklearn.metrics import accuracy_score
print("Accuracy is : ", accuracy_score(y_test, pred)*100)
```

**Visualization**

```python
sns.pairplot(df)
```

```python
# visualizing the graph
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(clf,filled=True)
```

**Prediction of right class of new data**

```python
# Testing new data
New_Data = ['Iris-Setosa', 'Iris-Viginica', 'Iris-versicolor']
test = np.array([6.3,3.3,6.0,2.5])
test = test.reshape(1,-1)
pred = clf.predict(test)
print("Input data: ",test)
print("Category: ",New_Data[int(pred[0])])
```
